# **Format Time from Seconds**

# **Instructions**

You are given a total number of seconds. Convert it to hours, minutes, and seconds and print it in the format hh:mm:ss.



Sample Input:

3661



# **Sample Output:**

01:01:01

```python
# Taking input
total_seconds = int(input())

# Write your code here
output =  ('0' * (( total_seconds // 3600) < 10)) + str( total_seconds // 3600) + ':' + ('0' * (( ( total_seconds % 3600) // 60 ) < 10)) +  str( ( total_seconds % 3600) // 60 ) + ':' + ('0' * (( total_seconds % 60 ) < 10)) +  str( total_seconds % 60 )

# Print the output
print(output)
```
