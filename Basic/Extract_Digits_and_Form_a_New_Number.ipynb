# **Extract Digits and Form a New Number**

# **Instructions**

You are given a 3-digit number. Extract the first and last digits, then form a new 2-digit number where the first digit is the last digit of the input and the second digit is the first digit of the input.



Sample Input:

781



# **Sample Output:**

**17**

```python
# Taking input
num = int(input())

# Write your code here
new_number = ( ( num % 100 ) % 10 ) * 10 + ( num // 100 )
#new_number = ( num % 100 ) % 10
#new_number = num // 100
# Print the output
print(new_number)
```
