# **Write a Python program to take two integers x and y as input from a user and solve the following expression: (x^y+(x+y)^(x-y))**



Note: ^ here specifies "to the power of". For example: 3^3 = 27.



Sample Input:

4

2



**Sample Output:**

**52**

```python
# Taking input
x = int(input())
y = int(input())

# Write your code here
output = ( x ** y + ( x + y ) ** ( x - y ))
# Print the output
print(output)
```
